```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#filename = "/content/test_2 (1).csv"
url="https://raw.githubusercontent.com/dannyvas23/AnalisisDataset/main/star_classification.csv" #importado desde el GitHub
```

```python
 #df = pd.read_excel(filename)
df= pd.read_csv( url , sep=',' )
df.head(5)
```

```python
print(df.dtypes)
```

```python
df['obj_ID'] = df['obj_ID'].astype('int64')
df['spec_obj_ID'] = df['spec_obj_ID'].astype('int64')
#df['redshift'] = df['redshift'].astype('int64')
```

```python
df.head(5)
```

```python
t = {'GALAXY':'?','QSO':'?','STAR':'?'}
df['class_value'] = df['class'].map(t)
df.head(5)
```

```python
train, test = train_test_split(df, test_size = 0.20) #division
```

```python
test.head(5)
```

```python
train_reset=train.set_index('obj_ID')
test_reset=test.set_index('obj_ID')
```

```python
test_reset.head(5)
```

```python
print("Ejemplos usados para entrenar: ", len(train_reset))
print("Ejemplos usados para test: ", len(test_reset))
```

```python
ruta = "/content/test_2.csv" #ruta y nombre de archivo
test_reset.to_csv(ruta)
```
